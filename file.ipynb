```python
import pandas as pd
def karl_pearson_skewness(data):
    mean = data.mean()
    median = data.median()
    std_dev = data.std()
    skewness = (3 * (mean - median)) / std_dev
    return skewness
```

```python
import numpy as np
def karl_pearson_correlation(x, y):
    mean_x = np.mean(x)
    mean_y = np.mean(y)
    covariance = np.sum((x - mean_x) * (y - mean_y))
    std_dev_x = np.sqrt(np.sum((x - mean_x) ** 2))
    std_dev_y = np.sqrt(np.sum((y - mean_y) ** 2))
    correlation_coefficient = covariance / (std_dev_x * std_dev_y)
    return correlation_coefficient
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
correlation = karl_pearson_correlation(x, y)
print("Karl Pearson correlation coefficient:", correlation)
```
